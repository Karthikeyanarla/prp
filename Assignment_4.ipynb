```python
# Importing all the necessary libraries
import networkx
import numpy as np
import scipy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy.random as rd
import cv2
import matplotlib.cm as cm
```

```python
# No of Colors or number of circle markers on the image for segmentation
color_count = 4

# Reading the input images

image = np.array(cv2.imread("a4_data/3.png"))
# image = cv2.resize(image, (500, 500))
count = 0

seeds = [[] for i in range(color_count)]

# Taking the following colors for the segmentation
colors = [(0,0,255), (128,0,0), (128,0,128), (255,165,0)]
# print(colors)


# Setting default(current) pointer or marker to 1 or maroon
pointer = 1
marks_updated = False


# Callback function for storing the seeds indices or the pixels marked
def mouse_callback(event, x, y, flags, param):
    global marks_updated
    
    if event == cv2.EVENT_LBUTTONDOWN:
        print(x,y)
        try:
            seeds[pointer].append((y,x))
            # Circle marker
            cv2.circle(image, (x, y), 2, colors[pointer], -1)

            marks_updated = True
            count += 1
 
        except:
            0

cv2.namedWindow('Input Image')
cv2.setMouseCallback('Input Image', mouse_callback)
 
while True:
    cv2.imshow('Input Image', image)
    key = cv2.waitKey(1)
    if key == 27:
        break
    elif key == ord('c'):
        seeds = [[] for i in range(color_count)]
    elif key > 0 and chr(key).isdigit():        
        pointer = int(chr(key))
    marks_updated = False
print(seeds)
cv2.destroyAllWindows()

# Total seeds count
seed_count = 0

# Seeds count of respective colors
color_seed_count = []
for i in range(color_count):
    seed_count += len(seeds[i])
    color_seed_count.append(len(seeds[i]))
```

```python
# Getting the neighbours of the each pixel for the construction of the graphs
def neighbours_pixels(x, y, img_shape):
    neighbours = []
    for pixel in [(x, y-1), (x-1, y)]:
#         print([(x, y-1), (x-1, y)])
        u, v = pixel
        if u >= 0 and v >= 0:
            neighbours.append((u, v))
    return neighbours
```

```python
# Function to calculate the weight given in the research paper
def weight(gi, gj, beta):
    gi = gi/100
    gj = gj/100
    
    wij = -(gi-gj)*(gi-gj)*beta
    return np.exp(wij)
```

```python
# Calculating the weighted graph
def weighted_graph_construction(graph,image,dimension):
    pair = {}
    for x in range(dimension[0]):
        for y in range(dimension[1]):
            neighbour_pixel_list = neighbours_pixels(x,y,img_shape)
            for pixel in neighbour_pixel_list:
                input_pixel, neighbour_pixel = image[x][y], image[pixel[0]][pixel[1]]
                w = weight(input_pixel, neighbour_pixel, 100)
                node1 = int(str(x) + "000" + str(y))
                node2 = int(str(pixel[0])+ "000"+ str(pixel[1]))
                pair[node1] = (x,y)
                graph.add_edge(node1, node2, weight=w)
    return graph,pair
```

```python
# Sorting the lists
def sorted_nodes_list(nodes_list, seeds_list):
    print(seeds_list)
    for seed in seeds_list[::-1]:
        for j in seed: 
            seed_node = int(str(j[0])+"000"+str(j[1]))
            nodes_list.insert(0, nodes_list.pop(nodes_list.index(seed_node)))
    return nodes_list
```

```python
image = cv2.imread("a4_data/3.png")
graph = networkx.Graph()
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img_shape = gray_image.shape
print(gray_image.shape)

print("building graph")
graph,pair = weighted_graph_construction(graph,gray_image,img_shape)
print("ordering nodes")
sorted_nodes = sorted_nodes_list(list(graph), seeds)
print("constructing laplacian")
laplacian_matrix  = networkx.laplacian_matrix(graph,nodelist = sorted_nodes, weight = "weight")    
laplacian = laplacian_matrix[seed_count: ,seed_count:]
laplacianb = laplacian_matrix[seed_count:, :seed_count]


print(laplacianb.shape)
print(laplacian.shape)

print('solving linear systems')

probability = []
for seed_index in range(color_count):
    print(f'Solving system {seed_index+1} out of {color_count}')
    seeds_vector = [0] * seed_count
    for i in range(sum(color_seed_count[:seed_index]),sum(color_seed_count[:seed_index])+color_seed_count[seed_index]):
        seeds_vector[i] = 1
    print(seeds_vector)
    probability_temp = scipy.sparse.linalg.spsolve(laplacian, -laplacianb@seeds_vector)
    probability.append(probability_temp)
print(np.array(probability).shape)
```

```python
index_array = []
for i in range(img_shape[0]):
    for j in range(img_shape[1]):
        index_array.append((i,j))
for k in seeds:
    for j in k:
        index_array.pop(index_array.index(j))

original_image = cv2.imread("a4_data/3.png")
for j in range(len(index_array)):
    new_list = []
    for k in range(4):
        new_list.append(probability[k][j])
    image[index_array[j][0]][index_array[j][1]] = colors[new_list.index(max(new_list))]


plt.subplot(1,2,1)
plt.imshow(original_image)
plt.subplot(1,2,2)    
plt.imshow(image)
```
